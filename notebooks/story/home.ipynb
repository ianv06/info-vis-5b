# Introduction
What factors help improve academic performance has always been a crucial question of generations of students and also researchers in the education field (e.g. https://doi.org/10.1001/jamapediatrics.2019.3176, https://doi.org/10.1136/bjsports-2017-098136, https://doi.org/10.1007/s10648-020-09585-3).

In some perspectives, a successful study method is investing most of study time per day without any distraction from social activities, gaming or entertainment. Previous study shows that study habits, skills, and attitudes can predict student academic performance, meaning that having serious learning behaviors for everyday plays an important role in grades (https://doi.org/10.1111/j.1745-6924.2008.00089.x). 

However, nowadays, student life is not just studying. They also spent time socializing and participating in other physical or extracurricular activities. Studying in a strict way might not be the most effective method to acquire high scores. https://doi.org/10.3390/ijerph17165972 found that sports participation may benefit students’ academic achievement. Besides, according to https://doi.org/10.1002/pits.10136 findings, participating in extracurricular activities in a structured way can also enhance school performance of students. 

Therefore, this data story aims to investigate whether living chill or living strictly would be more beneficial to students’ study, and the role of external variables such as stress level and parental supervision in the relationship between type of lifestyle and academic performance. Four datasets shared on Kaggle were used to structure the story: “Student lifestyle dataset” (https://doi.org/10.1002/pits.10136
https://www.kaggle.com/datasets/steve1215rogg/student-lifestyle-dataset), “Gaming vs Academic Performance” (https://www.kaggle.com/datasets/aiexplorer77/gaming-vs-academic-performance), “Student Alcohol Consumption” (https://www.kaggle.com/datasets/uciml/student-alcohol-consumption) and “Student Habits vs Academic Performance” (https://www.kaggle.com/datasets/jayaantanaath/student-habits-vs-academic-performance).
