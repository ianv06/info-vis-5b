# Introduction
What factors help improve academic performance has always been a crucial question of generations of students and also researchers in the educational field (e.g. https://doi.org/10.1001/jamapediatrics.2019.3176, https://doi.org/10.1136/bjsports-2017-098136, https://doi.org/10.1007/s10648-020-09585-3).

In some perspectives, a successful study method is investing most of your daily time to study, without any distractions from social activities, gaming or entertainment. Previous studies show that having serious learning behaviors for every day plays an important role in grades (https://doi.org/10.1111/j.1745-6924.2008.00089.x). However, student life is not just studying. They also spend time socializing and participating in other physical or extracurricular activities. Studying in a strict way might not be the most effective method to acquire high scores. Studies found that sports participation and extracurricular activities may also benefit students’ academic achievement (https://doi.org/10.3390/ijerph17165972,  https://doi.org/10.1002/pits.10136). 

Therefore, this data story aims to investigate whether living a balanced or strict lifestyle would be more beneficial to students’ study habits, and the role of external variables such as study hours and gaming in the relationship between type of lifestyle and academic performance. Specifically, two central perspectives in this story are “Having a strict lifestyle (e.g. only focusing on studying and not participating in leisure activities) improves students’ academic performance.” and “Having a balanced lifestyle (e.g. combining studying and participating in leisure activities) improves students’ academic performance.”
