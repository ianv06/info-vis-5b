# Balanced Lifestyle

The balanced lifestyle is about combining studying and leisure activities. In this part we will show how alcohol consumption and gaming impact education performance. 

## Alcohol Consumption Intensity
The first argument that we use for the balanced lifestyle perspective is that alcohol consumption has a negative impact on the final grade of a student, this argument is based on the paper of https://doi.org/10.1016/j.jpubeco.2010.08.008. The paper found that increased alcohol consumption caused significant reductions in student achievement, particularly among high-performing students.
