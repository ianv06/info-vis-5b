# Strict lifestyle

Strict lifestyle means focusing more on studying and reducing leisure activities. In this part, we discuss how strict lifestyle affects academic performance.

## Study time

_Students spending more time studying tend to achieve higher academic performance._

```python
#| class: col-page-right
%run ../plots/study_time_plot.ipynb
```

 *Figure 1. Average Study Time by Grade Group Across Four Datasets*

Study time was chosen as the first indicator of strict lifestyle because it shows the students’ investment and prioritization of learning. Across four datasets, higher study time is consistently correlated with higher academic performance. However, the relationship is not similar across subjects, as shown in dataset 3. The reason is because of different learning demands of two subjects, as mathematics often requires more structured problem-solving tactics, while language relies heavily on continuous exposure, self-practicing and application in reality (https://doi.org/10.1002/curj.213, https://doi.org/10.1080/09588221.2020.1744666). However, study time does not determine academic success, it also depends on learning strategies, quality, attention, and motivation of students (https://doi.org/10.1007/s10212-012-0167-9).

## Absence
Students with fewer absences tend to achieve higher academic performance.


```python
#| class: col-page-right
%run ../plots/attendance_plot.ipynb
```

*Figure 2. Average Absences by Grade Group in Mathematics and Portuguese* 

Absence was selected as another indicator of strict lifestyle, as it reflects students’ academic discipline. In both Mathematics and Portuguese, students in high grade group show fewest average absences, while medium and low grade groups are absent more frequently. This is supported by https://doi.org/10.3102/0034654310362998 that absenteeism can lead to lower grades, as missing classes reduces learning opportunities and classroom engagement. 

## Extra classes

_Students who get higher grades tend to combine intensive study time with extra classes_

```python
#| class: col-page-right
%run ../plots/extra_class_plot.ipynb
```

*Figure 3. Average Final Grade by Study Time and Paid Extra Classes*

Extra classes were included as another indicator of strict lifestyle because it reports students’ additional investment beyond mandatory classes. However, this pattern differs by subject. For Mathematics, students attending extra classes showed higher grades as study time increased. This is because extra classes or study time may not fully explain mathematical grades; their performance may also depend on learning strategies and conceptual understanding (https://doi.org/10.14812/cuefd.1292823, https://doi.org/10.18009/jcer.844906). For Portuguese, students studying more without extra classes tend to achieve higher grades than those who combine both of them, while the latter showed a fluctuating pattern. This difference lies in the subject characteristics: mathematics requires teacher’s problem-solving methods and guidance (https://doi.org/10.12973/ejmste/79805, https://doi.org/10.14812/cuefd.1292823), while with foreign languages, self-study and practice in reality will be more effective in improving students’ language skills (https://doi.org/10.1017/s0272263116000280, https://doi.org/10.1080/09588221.2022.2055081).

### Stress level

_Despite possibly affecting mental wellbeing, strictly follow intense study schedule brings fruitful result_

```python
#| class: col-page-right
%run ../plots/pers1_subarg4.ipynb
```

*Figure 4 Grade Distribution by Stress Level*

Overall, data shows that high stress correlates with high grades. Since strict, intensive study schedules are a very plausible reason students report high stress (rather than e.g. financial stress or personal issues, given this is a student lifestyle dataset), this chart visually supports the idea that strict lifestyles pay off academically. This phenomenon can be explained by the concept of eustress, which is a type of positive stress that result in positive outcome, which stems from the Yerkes-Dodson Law stating that ‘‘increasing stress is beneficial to performance until some optimum level is reached…’’ (https://doi.org/10.1108/02683940310502412) However, it is worth noting that grades spread quite widely in the high stress group, which is shown more clearly in the with-datapoints version, thus the pattern may not apply universally every time.
