### Design justification 1
The primary mark used in this visualization is area, this is represented by rectangular bars. This mark was chosen because bar-charts are highly effective at showing proportions and frequencies due to it being represented by the length and area of each bar on the chart. We didn't use point or line marks because these are mostly relevant for visualisations that represent single observations or continuous trends which isn't the case for us.

As visual channel we mainly used position for the x- and y-axis which range from 1 to 5 and 0 to 50 percent respectively. This is done because numerical values from 1 to 5 are very easy to interpret for users and percentages are great at representing quantitative comparison. in addition, we also used the channels size and color. Size is encoded by the height of the bar that can be seen on the visualization and color is encoded by giving both datasets (math and Portuguese) a different color to make them easily distinguishable for the user. lastly, we didn't use other channels such as value, texture, orientation, and shape because they would only over complexify the visualization and wouldn't add additional meaningfulness for the user. 

The visualization contains several data types, it primarily consists of ordinal data, this is because the alcohol consumption is measured in levels that can be ordered from 1 to 5 to represent ordered categories ranging from low to high alcohol consumption. This visualization also uses nominal and ratio data, the nominal data is represented by the labels math and Portuguese in de visualization in addition of different colors, the ratio data can be found on the y-axis where we used percentages to represent the bar length. Percentages are ratio data, because they have a meaningful zero point and allow comparisons.

### Design justification 2

The visualizations were designed to show the relationship between grades, reaction time, and gaming hours as clearly as possible. In the first figure, a 3D scatter plot is used so that all three variables can be displayed simultaneously: grades on one axis, reaction time on another, and gaming hours on the third. The axes are explicitly labeled, including the note that lower reaction times mean faster responses. Data points are colored by grade category using an intuitive traffic‑light palette (green for good, orange for sufficient, red for fail), which makes differences in performance easy to recognize at a glance and also keeps the categories visually consistent across figures. The amount of data points is reduced to improve visibility by taking a sample of 500 of the rows. This keeps hte same results, but make the graph more legible.

The second figure complements this view by combining several layers in a 2D layout. The central scatter plot shows grades against reaction time, with point size encoding gaming hours so that the third variable is still visible without requiring 3D rotation. Above this scatter, violin plots summarize the distribution of gaming hours for the different grade categories, making it easier to compare typical gaming behaviour between high‑ and low‑performing students rather than focusing on individual outliers. On the right, boxplots of grades by category display medians and variability, reinforcing the separation between good, sufficient, and failing groups. The same color scheme as in the first plot is reused so that the legend only needs to be learned once and the eye can move between figures without confusion.
