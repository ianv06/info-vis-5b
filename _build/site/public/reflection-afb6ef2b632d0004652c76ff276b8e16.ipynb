## Perspective 1
### Design Justifications
#### Figure 1
We used bar charts to compare the average study time across low, medium, and high grade groups. Four datasets were included to show consistent patterns of the data that high grade groups show more investment on their study time, not to compare between datasets from different schools. To improve comparability, study time was rescaled into the same 0-10 scale, clearly showing how students’ study time within the same time frame affected their scores. In the Student Alcohol Consumption dataset, the dataset was appended into a single table, with new column called “Subject” as the legend to distinguish two subjects. 
#### Figure 2
We used grouped bar charts to compare the average absent number of Mathematics and Portuguese students across grade groups. The x-axis indicates the students’ grade group, while y-axis shows the average absences. The legend was used to encode two subjects, helping audience distinguish whether the absence pattern differs between two subjects.

#### Figure 3
A line chart was used to illustrate how the relationship between combined study time and average final grade differs depending on whether students attended extra classes or not. Study time is shown on the x-axis, average final grade is represented as the y-axis, and legend color distinguishes students who attended extra classes and those who did not attend, making differences in trend to easily compare.

#### Figure 4
We believe median is the most suitable statistic to represent the typical performance of the groups because the size of different stress groups are not equal. Extreme values are more likely to occur by chance in larger groups and dragging the mean around. From that, boxplot is chosen as it not only displays the median effectively but also shows the distribution of students’ grades in each group. Two versions of graphs (with and without datapoints) are available for readers who want to inspect the spread of outliers in detail. Contrast of colors is utilized: darker shade of red delivers an increasing level of stress.

## Perspective 2
### Design Justifications

#### Figure 5
Here a histogram is used as the main mark to compare alcohol consumption levels and percentages. Position, height, and color encode the values, with separate colors for the Mathematics and Portuguese datasets. The data includes ordinal, nominal, and ratio variables, which makes the bar chart a suitable choice for clear category comparison.

#### Figure 6
Figure 6 uses boxplots to show the distribution of final grades across alcohol consumption levels. Position, size, and color are used to display the values and distinguish the two datasets. The plot also includes outliers, median values, and spread, giving a fuller view of how grades vary across categories.

#### Figure 7

Figure 7 shows the relationship between grades, reaction time, and gaming hours in a 3D scatter plot, with each axis clearly labeled and a note that lower reaction times indicate faster responses. The points are colored by grade category using a traffic-light palette, and only 500 rows are randomly sampled to keep the figure easier to read.

#### Figure 8

Figure 8 builds on Figure 7 by using a 2D scatter plot to show the relationship between grades and reaction time, while the violin plots highlight how gaming hours are distributed across grade categories. The consistent color scheme preserves visual continuity across the figures, and the random sample of 800 rows was used to keep the visualization clear and readable.

## Peer feedback
On June 23rd 2026, our group presented the story draft to the class and received some peer feedback, which mainly focused on strengthening the interpretation of our visualizations and showing the distinct comparisons across datasets more clearly: 
- Normalize the study time into the same scale in figure 1.
- Focus the study discipline explanation on Dataset 3 (absence level), as the other datasets show similar attendance percentages across groups.
- Clarify and emphasize the justification for stress level in the strict lifestyle.
- Simplify the alcohol consumption intensity by reducing the repeated patterns in the bar chart.
- Remove the boxplot from grades versus reaction time with a gaming scatter plot.

Following the feedback, our team reviewed the work carefully and made changes accordingly.
