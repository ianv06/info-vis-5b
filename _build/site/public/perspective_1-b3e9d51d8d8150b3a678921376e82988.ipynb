## Study time

_Students spending more time studying tend to achieve higher academic performance._

```python
#| class: col-page-right

%run ../plots/study_time_plot.ipynb
```

 *Figure 1. Average Study Time by Grade Group Across Four Datasets*

### Interpretation
The figure indicates that students who invest more study time tend to achieve higher grades. Across four datasets, the high grade group shows the highest average study time, while the low grade group consistently spends the least one. Additionally, the Student Alcohol Consumption dataset reveals a small and consistent difference between subjects. Students studying Mathematics report slightly more study time than those studying Portuguese. This is due to different learning demands of two subjects, as mathematics often requires more structured problem-solving tactics, while language relies heavily on continuous exposure, self-practicing and application in reality (https://doi.org/10.1002/curj.213, https://doi.org/10.1080/09588221.2020.1744666). 

Although higher study time is correlated with higher academic performance, showing across four datasets, study time alone does not fully explain students’ grades. It can still depend on learning strategies, quality, attention, and motivation of students (https://doi.org/10.1007/s10212-012-0167-9). Therefore, study time can be described as one of the important indicators of academic investment rather than a sole determinant.

*Figure 3. Average Final Grade by Study Time and Paid Extra Classes*

### Interpretation
The visualization indicates that the relationship between extra classes and study time, and average of final grade differs between mathematics and Portuguese. This is because whether math students attend extra classes or spend more time studying may not fully explain their grades; their performance may also depend on math-solving tactics, learning strategies, and conceptual understanding ( https://doi.org/10.14812/cuefd.1292823, https://doi.org/10.18009/jcer.844906). In terms of Portuguese, students who spend more time studying without extra classes tend to achieve higher grades than those who combine both of them. Interestingly, the group attending extra classes displayed a fluctuating pattern, with higher grades when studying 2 hours, but falling noticeably at 3 hours and recovering slightly at 4. This difference lies in the characteristics of the subject: mathematics requires problem-solving methods and teaching techniques from teachers, as well as concentration and comprehension from students. (https://doi.org/10.12973/ejmste/79805,  https://doi.org/10.14812/cuefd.1292823), while with foreign languages, self-study and practice in reality will be more effective in improving students' language skills and vocabulary ( https://doi.org/10.1017/S0272263116000280, https://doi.org/10.1080/09588221.2022.2055081).


### Stress level

_Despite may affect mental wellbeing, strictly follow intense study schedule brings fruitful result_

```python
#| class: col-page-right


%run ../plots/pers1_subarg4.ipynb
```

#### Interpretation

In the without-datapoints version of the graph, it is quite clearly shown that the interquartile range of the grades of students in the high stress group spread around the highest interval. This supports that students with high stress typically score higher than those with moderate and low stress on average. Also, as the stress level is higher, the median of grades in each group is also higher. Because GPA grade here is the average of many courses, a difference of 0.1 is already notable. This phenomenon can be explained by the concept of eustress, which is a type of positive stress that result in positive outcome, which stems from the Yerkes-Dodson Law, which states that ‘‘increasing stress is beneficial to performance until some optimum level is reached…’’ (https://doi.org/10.1108/02683940310502412) 
However, it is worth noting that grades spread quite widely in the high stress group, which is shown more clearly in the with-datapoints version. There is a group with high stress but relatively low grade (approximately as low as the low stress group), thus the pattern may not apply universally all the time. 
Overall, data shows that high stress correlates with high grades. Since strict, intensive study schedules are a very plausible reason students report high stress (rather than e.g. financial stress or personal issues, given this is a student lifestyle dataset), this chart visually supports the idea that strict lifestyles pay off academically.
