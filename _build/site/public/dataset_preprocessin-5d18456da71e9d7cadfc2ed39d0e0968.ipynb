# Dataset and Preprocessing

In assignment 1, two of team members interested in different factors affect students’academic performance. Therefore, when choosing the topic, our team decided to collect four datasets from Kaggle in the purpose of discovering how different factors influence students’ academic performance.  Although each dataset contains different variables and was collected from different populations, they all focus on the relationship between students’ study time and their academic performance. This allows us to test the same topic with multiple perspectives and datasets as evidence. Based on the datasets, we developed two different arguments to compare how different student lifestyles (strict versus chill lifestyles) impact on academic performance. Each argument will be supported by several substantiations that investigate specific factors such as study time, stress level, parental education, social time, etc., and different leisure activities. 





# Preprocessing


## Dataset 1: Student lifestyle and academic performance
The first dataset used in this project is about Student Lifestyle (https://www.kaggle.com/datasets/steve1215rogg/student-lifestyle-dataset), providing the information about how different factors in student lifestyle (such as students' daily habits across study, extracurriculars, sleep, socializing, stress levels, CGPA, and physical activities) can influence their academic performance (represented by GPA) and student well-being. This dataset contains 2000 records and 8 variables. In this project, we decided to use variables such as study hours, social hours, physical activity, stress level, GPA. For data preprocessing, we first checked the dataset structure by using pandas.DataFrame.head (n=5) to preview the first five rows, basic descriptive statistics through the kaggle dataset column, then checking variable name, data types, and missing values. Variables that were not relevant to our research were deleted. Second, for clearer visualization and consistency between datasets with different grading systems (e.g., GPA, final grade), we encoded students grade variables into categorical variable called Grade Group. Instead of using fixed score thresholds, we used a percentile tertile grouping (pandas.qcut) to divide students grade into three approximately equal-sized groups: Low, Medium, High. This allows us to ensure comparable proportions of grading across datasets. Finally, instead of merging dataset, we used each dataset to answer specific substantiation within our broader argument. This approach allows us to build more coherent storyline and show consistent patterns, while avoiding incorrect comparison between datasets with different samples and measurement scales.


## Dataset 4: Student Habits vs Academic Performance

The last dataset that we utilized is “Student Habits vs Academic Performance” (https://www.kaggle.com/datasets/jayaantanaath/student-habits-vs-academic-performance/data), containing habit-related variables such as sleep hours, diet quality, parttime job participation, etc. beside demographics records and their according exam grade, which represent academic performance as mentioned. Overall, this dataset has 1000 cases in total and 16 variables. Variables from this dataset that are being used in this study includes gender, study hour, exam grade, attendance, xxx.

To preprocess this dataset, we first removed duplicate data and those that are irrelevant. Then, we filtered out outliers by applying data onto the interquartile range (IQR); outliers removed were the values that lie outside [Q1 - 1.5IQR, Q3 + 1.5IQR], with IQR = Q3 - Q1. Lastly, we dealt with both numerical and categorical missing values in the dataset. Specifically, for 10 numerical variables, we replaced the spare space with the median value, while for the values from 6 categorical (string) variables, we filled in with mode.
