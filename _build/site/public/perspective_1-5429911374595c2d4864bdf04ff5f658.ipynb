## Extra classes

_Students who get higher grades tend to combine intensive study time with extra classes_

```python
%run plots/extra_class_plot.ipynb
```

### Interpretation
In this substantiation, the line chart was used to show the academic investment represented by paid extra classes. The visualization indicates that the relationship between extra classes and study time lead to higher grades of students. Students who combine intensive study time with additional classes often achieve higher grades. However, the pattern does not show the same regarding Math and Portuguese. When it comes to Math students, those who joined academic support classes tend to have gradually increased grades even if they have low or high study time. In contrast, Math students who do not attend extra classes will have lower grades if they spend less time studying, but higher grades when they invest enough time for studying. Therefore, whether math students attend extra classes or spend more time studying may not fully explain their grades; their performance may also depend on problem-solving skills, learning strategies, and conceptual understanding ( https://doi.org/10.14812/cuefd.1292823, https://doi.org/10.18009/jcer.844906). In terms of Portuguese, students who spend more time studying without extra classes tend to achieve higher grades than those who combine both of them. This difference lies in the nature of the subject. Mathematics requires teaching and various learning methods from the teacher (https://doi.org/10.12973/ejmste/79805,  https://doi.org/10.14812/cuefd.1292823), while with foreign languages, self-study and practice will be more effective in improving students' language skills and vocabulary ( https://doi.org/10.1017/S0272263116000280, https://doi.org/10.1080/09588221.2022.2055081).

### Data justification
The line charts are used to describe the relationship between combined study time and extra classes and average final grade of students. For both charts, the final grade is interval variable (y-axis), while study time is ordinal (x-axis). The two colored lines show students who attended extra classes and those who did not attend, coding as the legend of the graph. This design is useful because it shows the difference between extra-class groups, and the patterns change across different levels of study time.
