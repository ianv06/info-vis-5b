# Dataset and Preprocessing

## Datasets

We chose four datasets from Kaggle, which all include variables about lifestyle and education.

**Dataset 1: Student lifestyle and academic performance** (https://www.kaggle.com/datasets/steve1215rogg/student-lifestyle-dataset) 

**Dataset 2: Gaming dataset** (https://www.kaggle.com/datasets/aiexplorer77/gaming-vs-academic-performance)

**Dataset 3: Alcohol consumption dataset** (https://www.kaggle.com/datasets/uciml/student-alcohol-consumption). 

**Dataset 4: Student Habits vs Academic Performance** (https://www.kaggle.com/datasets/jayaantanaath/student-habits-vs-academic-performance/data)

## Variable description
Variables used in the data story are listed and categorized below.
Dataset 1 (D1) - Student lifestyle and academic performance:
- `Stress level` (ordinal): Level of stress reported in 3 levels - low, moderate, high
- `GPA` (ratio): Average grade on a 4.0 max scale

Dataset 2 (D2) - Gaming vs Academic Performance:
- `Gaming hours` (ratio): Average gaming hours per day
- `Reaction time` (ratio): Average response time
- `Grades` (ratio): Final grade score from 0-100 scale
- `Grade category` (nominal): classifies students’ performance into three ordered levels: fail, sufficient, and good. 

Dataset 3 (D3) - Student Alcohol Consumption:
- `Study time` (ordinal): Weekly study time in four categories (1 = <2 hours, 2 = 2-5 hours, 3 = 5-10 hours, 4 = > 10 hours)
- `Final grade`(ratio): Final grade (G3) on 0-20 scale
- `Weekday alcohol consumption`(ordinal): Alcohol consumption during weekdays rated from very low (1) to very high (5)
- `Weekend alcohol consumption` (ordinal): Alcohol consumption during weekends rated from very low (1) to very high (5)
- `Absences` (ratio): Number of school absences
- `Paid extra classes` (nominal): Whether the student attended paid extra educational support (Yes/No).
- `Course` (nominal) whether the observation is for math or Portuguese classes, these are separate datasets.

Dataset 4 (D4) - Student Habits vs Academic Performance:
- `Study hours per day` (ratio): Average daily study time.
- `Exam score` (ratio): Final examination score.


## Preprocessing
To preprocess datasets above, we first inspect the descriptive details of each dataset and remove irrelevant variables. Then, we checked for missing values and duplicate records which we removed. To handle missing values, we used median imputation for numerical variables and mode imputation for categorical variables. Besides, we also check for outliers. Lastly, we removed illogical datapoints, for example those who have grades above 100 (the maximum of grading scale used) in dataset 3. 

1. Specific preprocessing attempts in perspective 1

In dataset 1, we encoded students’ grade variables into categorical variables called Grade Group using percentile tertile grouping (pandas.qcut), which divides students’ grade into three approximately equal-sized groups: Low, Medium, High.

2. Specific preprocessing attempts in perspective 2

In dataset 3, we generalised the grades from 1-20 to 1-10. Also, we decided to make a new column called ‘Alcohol’ to generalise alcohol consumption over the entire week by adding the columns workday alcohol (Walc) and weekend alcohol (Dalc) together and dividing the new number by two, so we remain with a 1 to 5 scale. 


## Normalization
$$
\text{Normalized Study Time}
=
\frac{x-x_{\min}}{x_{\max}-x_{\min}}\times10
$$

The x is the original study time, $ x_{\min} $ and $ x_{\max} $ are the minimum and maximum study time values within each dataset. This allows the comparison between datasets in study time to become meaningful and relative.
For example: In Dataset 4, study time ranges from 0 to 8.3 study hours per day. A student who studies 6 hours is rescaled as:
$$
\frac{6.0 - 0}{8.3 - 0} \times 10 = 7.23
$$

Then, a study time of 6 is represented as 7.23 on the 0-10 scale.
