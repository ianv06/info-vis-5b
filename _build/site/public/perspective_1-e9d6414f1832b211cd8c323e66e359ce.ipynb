## Study time

_Students spending more time studying tend to achieve higher academic performance._

```python
#| class: col-page-right

%run plots/study_time_plot.ipynb
```

 *Figure 1. Average Study Time by Grade Group Across Four Datasets*

### Interpretation
The figure indicates that students who invest more study time tend to achieve higher grades. Across four datasets, the high grade group shows the highest average study time, while the low grade group consistently spends the least one. Additionally, the Student Alcohol Consumption dataset reveals a small and consistent difference between subjects. Students studying Mathematics report slightly more study time than those studying Portuguese. This is due to different learning demands of two subjects, as mathematics often requires more structured problem-solving tactics, while language relies heavily on continuous exposure, self-practicing and application in reality (https://doi.org/10.1002/curj.213, https://doi.org/10.1080/09588221.2020.1744666). 

Although higher study time is correlated with higher academic performance, showing across four datasets, study time alone does not fully explain students’ grades. It can still depend on learning strategies, quality, attention, and motivation of students (https://doi.org/10.1007/s10212-012-0167-9). Therefore, study time can be described as one of the important indicators of academic investment rather than a sole determinant.

## Extra classes

_Students who get higher grades tend to combine intensive study time with extra classes_

```python
#| class: col-page-right


%run plots/extra_class_plot.ipynb
```

*Figure 3. Average Final Grade by Study Time and Paid Extra Classes*

### Interpretation
The visualization indicates that the relationship between extra classes and study time, and average of final grade differs between mathematics and Portuguese. The patterns are not fully linear, the trend fluctuated at some points. For Mathematics, students who did not attend extra classes experienced a slight decline in grades when studying 1 to 2 hours, followed by reaching a peak at 3 hours before dropping again at 4 hours. In contrast, students attending extra classes show a stable upward trend with increased average grades. This is because whether math students attend extra classes or spend more time studying may not fully explain their grades; their performance may also depend on math-solving tactics, learning strategies, and conceptual understanding ( https://doi.org/10.14812/cuefd.1292823, https://doi.org/10.18009/jcer.844906). In terms of Portuguese, students who spend more time studying without extra classes tend to achieve higher grades than those who combine both of them. Interestingly, the group attending extra classes displayed a fluctuating pattern, with higher grades when studying 2 hours, but falling noticeably at 3 hours and recovering slightly at 4. This difference lies in the characteristics of the subject: mathematics requires problem-solving methods and teaching techniques from teachers, as well as concentration and comprehension from students. (https://doi.org/10.12973/ejmste/79805,  https://doi.org/10.14812/cuefd.1292823), while with foreign languages, self-study and practice in reality will be more effective in improving students' language skills and vocabulary ( https://doi.org/10.1017/S0272263116000280, https://doi.org/10.1080/09588221.2022.2055081).
