## Perspective 1
### Design Justifications
#### Figure 1
We used bar charts to compare the average study time across low, medium, and high grade groups. Four datasets were included to show consistent patterns of the data that high grade groups show more investment on their study time, not to compare between datasets from different schools. To improve comparability, study time was rescaled into the same 0-10 scale, clearly showing how students’ study time within the same time frame affected their scores. In the Student Alcohol Consumption dataset, the dataset was appended into a single table, with new column called “Subject” as the legend to distinguish two subjects. 
#### Figure 2
We used grouped bar charts to compare the average absent number of Mathematics and Portuguese students across grade groups. The x-axis indicates the students’ grade group, while y-axis shows the average absences. The legend was used to encode two subjects, helping audience distinguish whether the absence pattern differs between two subjects.

#### Figure 3
A line chart was used to illustrate how the relationship between combined study time and average final grade differs depending on whether students attended extra classes or not. Study time is shown on the x-axis, average final grade is represented as the y-axis, and legend color distinguishes students who attended extra classes and those who did not attend, making differences in trend to easily compare.

#### Figure 4
We believe median is the most suitable statistic to represent the typical performance of the groups because the size of different stress groups are not equal. Extreme values are more likely to occur by chance in larger groups and dragging the mean around. From that, boxplot is chosen as it not only displays the median effectively but also shows the distribution of students’ grades in each group. Two versions of graphs (with and without datapoints) are available for readers who want to inspect the spread of outliers in detail. Contrast of colors is utilized: darker shade of red delivers an increasing level of stress.

## Perspective 2
### Design Justifications

#### Figure 5
The primary mark used in this visualization is area, represented by rectangular bars. Bar charts are effective for comparing proportions through bar height. Point and line marks were not used since the data does not represent individual observations or continuous trends.
The visualization uses position, size, and color as visual channels. Position represents alcohol consumption levels and percentages, size is encoded by bar height, and color distinguishes the Mathematics and Portuguese datasets.
The data consists of ordinal, nominal, and ratio types. Alcohol consumption levels are ordinal, the dataset categories are nominal, and the percentages are ratio data because they have a meaningful zero point and allow quantitative comparison.

#### Figure 6
The primary marks used in this visualization are area, point, and line. The boxes represent the distribution of final grades, points indicate outliers, and lines show the median and spread of the data.
The visualization uses position, size, and color as visual channels. Position represents alcohol consumption levels and final grades, size is encoded through the box and line heights, and color distinguishes the Math and Portuguese datasets.
The data consists of ordinal, nominal, and ratio types. Alcohol consumption levels are ordinal, the dataset categories are nominal, and final grades are ratio data because they have a meaningful zero point and allow quantitative comparison.

#### Figure 7

The visualizations were designed to show the relationship between grades, reaction time, and gaming hours as clearly as possible. In the first figure, a 3D scatter plot is used so that all three variables can be displayed simultaneously: grades on one axis, reaction time on another, and gaming hours on the third. The axes are explicitly labeled, including the note that lower reaction times mean faster responses. Data points are colored by grade category using an intuitive traffic‑light palette (green for good, orange for sufficient, red for fail), which makes differences in performance easy to recognize at a glance and also keeps the categories visually consistent across figures. The amount of data points is reduced to improve visibility by taking a sample of 500 of the rows. This keeps hte same results, but make the graph more legible.

#### Figure 8

The second figure complements this view by combining several layers in a 2D layout. The central scatter plot shows grades against reaction time, with point size encoding gaming hours so that the third variable is still visible without requiring 3D rotation. Above this scatter, violin plots summarize the distribution of gaming hours for the different grade categories, making it easier to compare typical gaming behaviour between high‑ and low‑performing students rather than focusing on individual outliers. On the right, boxplots of grades by category display medians and variability, reinforcing the separation between good, sufficient, and failing groups. The same color scheme as in the first plot is reused so that the legend only needs to be learned once and the eye can move between figures without confusion.

## Peer feedback
On June 23th 2026, our group presented the story draft to the class and received some peer feedback, which mainly focused on strengthening the interpretation of our visualizations and showing the distinct comparisons across datasets more clearly: 
- Normalize the study time into the same scale in figure 1.
- Focus the study discipline explanation on Dataset 3 (absence level), as the other datasets show similar attendance percentages across groups.
- Clarify and emphasize the justification for stress level in the strict lifestyle.
- Simplify the alcohol consumption intensity by reducing the repeated patterns in the bar chart.
- Remove the boxplot from grades versus reaction time with a gaming scatter plot.

Following the feedback, our team reviewed the work carefully and made changes accordingly.
