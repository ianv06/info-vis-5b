```python
import pandas as pd
import plotly.express as px

df3_mat = pd.read_csv("student-mat.csv", sep=",")
df3_por = pd.read_csv("student-por.csv", sep=",")
```

```python
df3_mat.head()
```

```python
df3_por.head()
```

```python

counts = df3_mat["Dalc"].value_counts(normalize=True).sort_index().reset_index()
counts.columns = ["Dalc", "Verhouding"]

fig = px.bar(
    counts,
    x="Dalc",
    y="Verhouding",
    text="Verhouding",
    labels={
        "Dalc": "Weekday alcohol consumption math",
        "Verhouding": "Percentage of students"
    }
)

fig.update_traces(texttemplate='%{y:.1%}', textposition='outside')
fig.update_yaxes(tickformat='.0%')

fig.show()
```

```python
counts = df3_mat["Walc"].value_counts(normalize=True).sort_index().reset_index()
counts.columns = ["Walc", "Verhouding"]

fig = px.bar(
    counts,
    x="Walc",
    y="Verhouding",
    text="Verhouding",
    labels={
        "Walc": "Weekend alcohol consumption math",
        "Verhouding": "Percentage of students"
    }
)

fig.update_traces(texttemplate='%{y:.1%}', textposition='outside')
fig.update_yaxes(tickformat='.0%')

fig.show()
```

```python
counts = df3_por["Walc"].value_counts(normalize=True).sort_index().reset_index()
counts.columns = ["Walc", "Verhouding"]

fig = px.bar(
    counts,
    x="Walc",
    y="Verhouding",
    text="Verhouding",
    #color="Walc",
    #color_continuous_scale="RdYlGn_r",
    labels={
        "Walc": "Weekend alcahol consumption Portuguese",
        "Verhouding": "Percentage of students"
    }
)

fig.update_traces(texttemplate='%{y:.1%}', textposition='outside')
fig.update_yaxes(tickformat='.0%')

fig.show()
```

```python
kleur_map = {
    1: "green",
    2: "limegreen",
    3: "gold",
    4: "orange",
    5: "red"
}

fig = px.box(
    df3_por,
    x="Walc",
    y="G3",
    color="Walc",
    color_discrete_map=kleur_map,
    labels={
        "Walc": "Weekend alcohol consumption",
        "G3": "Final grade Portuguese"
    },
    title="Final grade Portuguese per level of weekend alcohol consumption",
    category_orders={
        "Walc": [1, 2, 3, 4, 5]
    }
)

fig.show()
```
